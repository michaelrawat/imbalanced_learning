### Churn Data Set

```python
# Importing Churn data set
Churn = pd.read_csv(pathname + 'Datasets\\Churn.csv', encoding = 'utf-8')

# Checking data set was imported by printing first and last five records
print('\r')
print("Below shows the first five records in the data set:")
display(Churn.head())

# Checking the size of the data set
print('\r')
print("The data set has x records across y variables (x,y):")
print(Churn.shape)

######################### Numeric and Non-Numeric Variables #########################

# Describing all numeric variables
print('\r')
print("Below is information about every numeric variable present in the data set:")
display(Churn.describe(include=[np.number]))
print('\r')
print("Notice that "+str(Churn.describe(include=[np.number]).shape[1])+" columns of data contain numeric data.")

# Describing all non-numeric variables
if (Churn.describe(include=[np.number]).shape[1] <  Churn.shape[1]):
    print('\r')
    print("Below is information about every non-numeric variable present in the data set:")
    display(Churn.describe(exclude=[np.number]))
    print('\r')
    print("Notice that "+str(Churn.describe(exclude=[np.number]).shape[1])+" columns of data contain non-numeric data.")
else:
    print('\r')
    print("There are no non-numeric variables within the data set.")
    
######################### Missing Data #########################

# Creating a count of missing values in each column
missing_values = Churn.isnull().sum()

# Displaying the number of missing values in each column that contains a missing value
print('\r')
print("The variables below contain missing values:")
display(missing_values[missing_values>0])

######################### Re-locating and Converting Variables #########################

# Designating Target Variable
Target_Variable = 'Churn Flag'
Target_Series = Churn[Target_Variable]

# Removing target variable from current location and inserting in collum 0
del Churn[Target_Variable]
Churn.insert(0, Target_Variable, Target_Series)
        
# Showing the data frame after target variable has been moved to the front
print('\r')
print("Below shows the first five records in the data set after the target variable has been moved and converted:")
display(Churn.head())

# Converting predictor into numeric variable for learning algorithms later
numeric_encoder = preprocessing.LabelEncoder()
Churn['AON'] = numeric_encoder.fit_transform(Churn['AON'])

######################### Conducting Min-Max Standardization #########################

# Creating a list of numeric variable names
List_of_Numeric_Variables = Churn.describe(include=[np.number]).columns.tolist()

# Creating a for loop to min-max standardize every numeric variable
for variable in List_of_Numeric_Variables:
    Churn[variable] = (Churn[variable]-Churn[variable].min())/(Churn[variable].max()-Churn[variable].min())

# Displaying updated numeric variable statistics to check min max standardization was successful
print('\r')
print("Below is an updated table of numeric variable statistics to demonstrate Min-Max Standardization has been successful:")
display(Churn.describe(include=[np.number]))

######################### Target Variable and Predictor Sets #########################

# Defining Target Variable Column
target_variable_column = Churn[Target_Variable]

# Checking the size of the target variable data set
print('\r')
print("The target variable data set (x,y) contains x records across y features:")
print(target_variable_column.shape)
print('\r')
print("Below is a list of all the variables present in the target variable data set:")
print('\r')
print(target_variable_column.name)

# Creating a data set contains all the predictor variables
predictor_variables_columns = Churn.drop([Target_Variable], axis=1) 

# Checking the size of the target variable data set
print('\r')
print("The predictor variables data set (x,y) contains x records across y features:")
print(predictor_variables_columns.shape)
print('\r')
print("Below is a list of all the variables present in the predictor variable data set:")
print('\r')
print(predictor_variables_columns.columns)

# Storing target variable and predictor variables to arrays for PCA
target_variable_array = target_variable_column.values
predictor_variables_array = predictor_variables_columns.values

######################### Exporting Cleaned Data Set for Analysis #########################

# Exporting cleaned data frame as csv
Churn.to_csv('Churn_Cleaned.csv', index=False)
```
